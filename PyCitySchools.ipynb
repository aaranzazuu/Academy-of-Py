```python
#Dependencies and setup
import pandas as pd
```

```python
# File to Load
school_data_to_load = "../pandas-challenge/PyCitySchools/Resources/04-Pandas_HW_Instructions_PyCitySchools_Resources_schools_complete.csv"
student_data_to_load = "../pandas-challenge/PyCitySchools/Resources/04-Pandas_HW_Instructions_PyCitySchools_Resources_students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Create a list of schools and Calculate Total Schools
schools_list = school_data_complete["school_name"].unique()
total_schools = len(schools_list)
```

```python
#Calculate Total Students
total_students = school_data_complete["Student ID"].count()
```

```python
#Calculate Total Budget
grouped_school = school_data_complete.groupby(["school_name"])
total_budget_unique = grouped_school["budget"].unique()
total_budget = "${0:,.2f}".format(int(total_budget_unique.sum()))
```

```python
#Average Scores
#Math
average_math = school_data_complete["math_score"].mean()
#Reading
average_reading = school_data_complete["reading_score"].mean()
```

```python
#Overall Passing Rate
overall_passing = "{0:,.2f}%".format((average_math + average_reading)/2)
```

```python
#Percentage Passing Scores
#Math
total_passing_math = school_data_complete[school_data_complete["math_score"] >= 70].count()
percentage_math = "{0:,.2f}%".format(total_passing_math["Student ID"]*100/total_students)
#Reading
total_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70].count()
percentage_reading = "{0:,.2f}%".format(total_passing_reading["Student ID"]*100/total_students)
```

```python
#District Summary
district_summary = pd.DataFrame({"Total Schools": total_schools, 
                                 "Total Students": total_students, 
                                 "Total Budget": total_budget, 
                                 "Average Math Score": average_math, 
                                 "Average Reading Score": average_reading, 
                                 "%Passing Math": percentage_math, 
                                 "%Passing Reading": percentage_reading, 
                                 "%Overall Passing Rate": overall_passing}, index=[0])
district_summary
```

```python
#Type per school
type_school = grouped_school["type"].unique()
#Remove brackets on results
type_school = type_school.str[0]
```

```python
#Totla students per school
total_students_byschool = grouped_school["student_name"].count()
total_students_byschool
#Create a Data Frame with Total Students by School and convert to Numeric. This will be used for the passing rates.
total_students_byschool_object = pd.DataFrame(total_students_byschool)
total_students_byschool_object
total_students_byschool_df = pd.to_numeric(total_students_byschool_object["student_name"])
```

```python
#Total Budget per school
total_budget_byschool = grouped_school["budget"].unique()
#Formatted total budget by school. Different variable to avoid affecting calculations that use the total budget by school
formatted_total_budget_byschool = total_budget_byschool.str[0].map("${:,.2f}".format)
```

```python
#Budget per student
budget_student_byschool = total_budget_byschool/total_students_byschool
formatted_budget_student_byschool = budget_student_byschool.str[0].map("${:,.2f}".format)
#Casting to integer for binning below
budget_student_byschool_int = budget_student_byschool.astype(int)
```

```python
#Average Math Score
average_math_byschool = grouped_school["math_score"].mean()
```

```python
#Average Reading Score
average_reading_byschool = grouped_school["reading_score"].mean()
```

```python
#Percentage of Students who passed math
#Count total students who passed math
total_passing_math_byschool = school_data_complete.loc[school_data_complete["math_score"] >= 70]
#Group count by school and convert to Data Frame
grouped_school_math = total_passing_math_byschool.groupby(["school_name"])["math_score"].count()
school_math_object = pd.DataFrame(school_math)
#Convert to int
school_math_df = pd.to_numeric(school_math_object["math_score"])
#Students who passed math divided by all students. Formatted as percentage
percentage_math_byschool = ((school_math_df*100)/total_students_byschool_df)
#Format as percentage. Created new variable for formatted percentage so I can use non-formatted variable in the overall pass rate
formatted_percentage_math_byschool = percentage_math_byschool.map("{:.2f}%".format)
```

```python
#Percentage of Students who passed reading
#Count total students who passed reading
total_passing_reading_byschool = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
#Group by count school and convert to Data Frame
group_school_reading = total_passing_reading_byschool.groupby(["school_name"])["reading_score"].count()
school_reading = pd.DataFrame(group_school_reading)
#Convert to int
school_reading_df = pd.to_numeric(school_reading["reading_score"])
#Students who passed reading divided by all students
percentage_reading_byschool = ((school_reading_df*100)/total_students_byschool_df)
#Format as percentage. Created new variable for formatted percentage so I can use non-formatted variable in the overall pass rate
formatted_percentage_reading_byschool = percentage_reading_byschool.map("{:.2f}%".format)
```

```python
#Overall Pass Rate
overall_passing_byschool = pd.DataFrame({"Passing Math": percentage_math_byschool, 
                                         "Passing Reading": percentage_reading_byschool}).mean(axis=1)
#Format as percentage. Created new variable for formatted percentage so I can use non-formatted variable in the binning
formatted_overall_passing_byschool = overall_passing_byschool.map("{:.2f}%".format)
```

```python
#School Summary
school_summary = pd.DataFrame({"School Type": type_school, 
                               "Total Students": total_students_byschool,
                                 "Total School Budget": formatted_total_budget_byschool, 
                               "Per Student Budget": formatted_budget_student_byschool,
                                 "Average Math Score": average_math_byschool, 
                                 "Average Reading Score": average_reading_byschool, 
                                 "%Passing Math": formatted_percentage_math_byschool, 
                                 "%Passing Reading": formatted_percentage_reading_byschool, 
                                 "%Overall Passing Rate": formatted_overall_passing_byschool})
school_summary
```

```python
#Top Performing Schools (By Passing Rate)
top_performing_schools = school_summary.sort_values("%Overall Passing Rate", ascending = False)
top_performing_schools.head()
```

```python
#Bottom Performing Schools (By Passing Rate)
bottom_performing_schools = school_summary.sort_values("%Overall Passing Rate", ascending = True)
bottom_performing_schools.head()
```

```python
#Create a dataframe for school_data_complete by grade. Averaged math score grouped by school. Averaged reading score grouped by school
#9th grade
ninth_grade = school_data_complete.loc[school_data_complete["grade"]=="9th"]
ninth_grade_byschool_math = ninth_grade.groupby(["school_name"])["math_score"].mean()
ninth_grade_byschool_reading = ninth_grade.groupby(["school_name"])["reading_score"].mean()

#10th grade
tenth_grade = school_data_complete.loc[school_data_complete["grade"]=="10th"]
tenth_grade_byschool_math = tenth_grade.groupby(["school_name"])["math_score"].mean()
tenth_grade_byschool_reading= tenth_grade.groupby(["school_name"])["reading_score"].mean()

#11th grade
eleventh_grade = school_data_complete.loc[school_data_complete["grade"]=="11th"]
eleventh_grade_byschool_math = eleventh_grade.groupby(["school_name"])["math_score"].mean()
eleventh_grade_byschool_reading = eleventh_grade.groupby(["school_name"])["reading_score"].mean()

#12th grade
twelfth_grade = school_data_complete.loc[school_data_complete["grade"]=="12th"]
twelfth_grade_byschool_math = eleventh_grade.groupby(["school_name"])["math_score"].mean()
twelfth_grade_byschool_reading = eleventh_grade.groupby(["school_name"])["reading_score"].mean()
```

```python
#Math Scores By Grade

math_scores_byschool = pd.DataFrame({"9th": ninth_grade_byschool_math, "10th": tenth_grade_byschool_math, 
                                    "11th": eleventh_grade_byschool_math, "12th": twelfth_grade_byschool_math})
math_scores_byschool
```

```python
#Reading Scores By Grade

reading_scores_byschool = pd.DataFrame({"9th": ninth_grade_byschool_reading, "10th": tenth_grade_byschool_reading, 
                                    "11th": eleventh_grade_byschool_reading, "12th": twelfth_grade_byschool_reading})
reading_scores_byschool
```

```python
#Creating a non-formatted version of school summary for binning below
non_formatted_school_summary = pd.DataFrame({"School Type": type_school, 
                                             "Total Students": total_students_byschool,
                                 "Total School Budget": total_budget_byschool.str[0], 
                               "Per Student Budget": budget_student_byschool_int,
                                 "Average Math Score": average_math_byschool, 
                                 "Average Reading Score": average_reading_byschool, 
                                 "%Passing Math": percentage_math_byschool, 
                                 "%Passing Reading": percentage_reading_byschool, 
                                 "%Overall Passing Rate": overall_passing_byschool})
non_formatted_school_summary
```

```python
#Binning schools by spending per student
bins = [0, 585, 615, 645, 675]
spending_ranges = ["<$585", "$585-615", "$615-645", "$645-675"]

bin_byspending = pd.cut(non_formatted_school_summary["Per Student Budget"], bins, labels = spending_ranges, include_lowest = True)
series_byspending = pd.Series(bin_byspending)

#School Summary with needed columns
school_summary_forbins = non_formatted_school_summary[["Average Math Score", "Average Reading Score", "%Passing Math", "%Passing Reading","%Overall Passing Rate"]]

#School Summary merged with Schools binned by spending per student 
spending_byschool_bins = pd.merge(school_summary_forbins, bin_byspending, on="school_name", how="outer")

#Rename of "Per Student Budget" for better comprehension
spending_byschool_bins = spending_byschool_bins.rename(columns={"Per Student Budget":"Spending Ranges (Per Student)"})
```

```python
#Scores by School Spending
#Group by spending ranges
grouped_spending_ranges = spending_byschool_bins.groupby(["Spending Ranges (Per Student)"]).mean()
grouped_spending_ranges
```

```python
#Binning schools by size
bins = [0, 1000, 2000, 5000]
size_ranges = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

bin_bysize = pd.cut(non_formatted_school_summary["Total Students"], bins, labels = size_ranges, include_lowest = True)
series_bysize = pd.Series(bin_bysize)
series_bysize

#School Summary merged with Schools binned by size
size_byschool_bins = pd.merge(school_summary_forbins, bin_bysize, on="school_name", how="outer")

#Rename of "Per Student Budget" for better comprehension
size_byschool_bins = size_byschool_bins.rename(columns={"Total Students":"School Size"})
```

```python
#Scores by School Spending
#Group by spending ranges
grouped_size_ranges = size_byschool_bins.groupby(["School Size"]).mean()
grouped_size_ranges
```
